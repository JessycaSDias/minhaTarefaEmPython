```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset
df = pd.read_excel(r"C:\Users\Jéssyca\Desktop\3°Período\Ciência dos dados para tomadas de decisões\default_of_credit_card_clients__courseware_version_1_21_19.xls")

# Exibir as primeiras 5 linhas para verificar os dados
print(df.head())

# Análise das colunas numéricas
numerical_columns = ['LIMIT_BAL', 'AGE', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 
                     'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 
                     'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6', 
                     'default payment next month']

# Gerar resumo estatístico das variáveis numéricas
print("\nResumo Estatístico das Variáveis Numéricas:")
print(df[numerical_columns].describe())

# Gerar gráficos de boxplot para as variáveis numéricas
plt.figure(figsize=(15, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(5, 5, i)  # Organizando os gráficos em uma grid 5x5
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot - {column}')
plt.tight_layout()
plt.show()

# Gerar histogramas para as variáveis numéricas
plt.figure(figsize=(15, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(5, 5, i)  # Organizando os gráficos em uma grid 5x5
    df[column].hist(bins=20, edgecolor='black')
    plt.title(f'Histograma - {column}')
plt.tight_layout()
plt.show()

# Análise das colunas categóricas
categorical_columns = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_1']

# Exibir a frequência de cada categoria nas colunas categóricas
print("\nFrequência das Categorias nas Colunas Categóricas:")
for col in categorical_columns:
    print(f"\nFrequência de {col}:")
    print(df[col].value_counts())

# Gerar gráficos de barras para visualizar a frequência das categorias
plt.figure(figsize=(15, 8))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(2, 2, i)  # Organizando os gráficos em uma grid 2x2
    sns.countplot(x=df[column])
    plt.title(f'Frequência - {column}')
plt.tight_layout()
plt.show()
```

```python
# Verificando se existem dados faltando em cada coluna
missing_data = df.isnull().sum()

# Verificando a porcentagem de dados faltando em cada coluna
missing_percentage = (missing_data / len(df)) * 100

# Exibindo os resultados
print("Número de dados faltando por coluna:")
print(missing_data)
print("\nPorcentagem de dados faltando por coluna:")
print(missing_percentage)

# Verificando se existe alguma coluna com dados faltando
if missing_data.any():
    print("\nExistem colunas com dados faltando.")
else:
    print("\nNão existem dados faltando no conjunto de dados.")
```
